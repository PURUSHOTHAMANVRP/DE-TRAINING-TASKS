```sql
CREATE PROCEDURE INSERT_EMPLOYEE
@FIRST_NAME STRING
@LAST_NAME STRING
@EMAIL STRING
@DEPARTMENT STRING
AS
BEGIN
SET NOCOUNT ON;

BEGIN TRY
INSERT INTO EMPLOYEES (FIRST_NAME, LAST_NAME, EMAIL, DEPARTMENT) VALUES (@FIRST_NAME, @LAST_NAME, @EMAIL, @DEPARTMENT)
END TRY

BEGIN CATCH
ERROR_NUMER() AS error_number;
ERROR_MESSAGE() AS error_message;
END CATCH
END;
```
